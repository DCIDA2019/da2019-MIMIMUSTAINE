```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(x):
    if x % 2 == 0:
        return x * x
    else:
        return np.log(x)
```

```python
xx = np.arange(1,50,3)
print(x)
#for i in range(len(x)):
#    f1(x[i])
```

```python
def f2(x):
    y = []
    for i in range(len(x)):
        if x[i] % 2 == 0:
            y.append(x[i] * x[i])
        else:
            y.append(np.log(x))
    
    return y
```

```python
f2(xx)
```

```python
f3 = np.vectorize(f2)
```

```python
def g(x):
    if x > 0:
        return np.cos(x)
    else:
        return np.sin(x)
g1 = np.vectorize(g)
```

```python
x2 = np.linspace(-2 * np.pi, 2 * np.pi, 50)
g1(x2)
```

```python
plt.plot(x2, g1(x2))
```

```python
y = np.zeros_like(x2)
w = x2 > 0
w
y[w] = g1(x2[w]) #Aplico g sobre los elementos de x en los que w es verdadero.
y[~w] = g1(x2[~w]) #Aplico g sobre los elementos de x en los que w es falso.
```

```python
y
```

```python
x2[w]  #Elementos en los que w es verdadero
```

```python
y2 = np.asarray([g(x) for x in x2])
%timeit y2
```

```python
def g2(x):
    y = np.zeros_like(x)
    for i in range(len(x)):
        if x[i] > 0:
            y[i] = np.cos(x[i])
        else:
            y[i] = np.sin(x[i])
    return y
```

```python
%timeit g2(x2)
```

```python
class Funciones(): #Clase que contiene una función que regresa 'Hola'.
    "Se definen mis funciones básicas, composición de una clase. A: amplitud de las funciones periódicas."
    def __init__(self, A, saludo = 'Hola'):
        self.A = A
        self.saludo = saludo
    def g(self, x):
        return self.A * self.A * np.cos(x), np.sin(x)
    
    def f(self): #Sólo requiere ser declarada sin argumentos de entrada, como funciones void en C.
        return self.saludo
    
    def f2(self,x):
        if x%2 == 0:
            return x * x
        else:
            if x > 0:
                return np.log(x)
```

```python
Funciones?
```

```python
x = Funciones(3, 'Hello')
```

```python
x.f()
```

```python
b = x.g(np.pi)
```
