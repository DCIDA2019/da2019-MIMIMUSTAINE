### <strong>Distribución uniforme en cubo</strong>

Para una distribución uniforme que ocupe el volumen de un cubo se hace uso de la función <i>np.random.uniform(límite inferior, límite superior, número de puntos)</i>.

```python
ax = plt.axes(projection='3d')

x = np.random.uniform(-1, 1, 2000)
y = np.random.uniform(-1, 1, 2000)
z = np.random.uniform(-1, 1, 2000)
dr = np.sqrt((x ** 2) + (y ** 2) + (z ** 2))

ax.scatter3D(x, y, z, color = 'purple', marker = '.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

### <strong>Distribución uniforme en esfera</strong>

Generando puntos en coordenadas esféricas y transformándolo a cartesianas se obtiene el siguiente resultado:

```python
ax = plt.axes(projection='3d')

r = np.random.uniform(0, 1, 5000)
phi = np.random.uniform(0, 2 * np.pi, 5000)
theta = np.random.uniform(0, np.pi, 5000)

#A cartesianas:
x = r * np.sin(theta) * np.cos(phi)
y = r * np.sin(theta) * np.sin(phi)
z = r * np.cos(theta)

ax.scatter3D(x, y, z, color = 'purple', marker = '.', s = 1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

Note que si se grafica en el espacio con ejes <strong>r, theta, phi</strong>, la distribución corresponde nuevamente a un cubo:

```python
ax = plt.axes(projection='3d')

ax.scatter3D(r, theta, phi, color = 'purple', marker = '.', s = 1)
ax.set_xlabel('r')
ax.set_ylabel('$\\theta$')
ax.set_zlabel('$\\phi$')
```

Y la producción de una esfera haciendo uso de las coordenadas cartesianas se realiza con una condición en la cual la norma del radio vector tiene que ser menor a un número específico:

```python
ax = plt.axes(projection='3d')
x = np.random.uniform(-1, 1, 5000)
y = np.random.uniform(-1, 1, 5000)
z = np.random.uniform(-1, 1, 5000)
v = np.sqrt(x ** 2 + y ** 2 + z ** 2)

x_1 = x[(np.sqrt(x ** 2 + y ** 2 + z ** 2) < 1)]
y_1 = y[(np.sqrt(x ** 2 + y ** 2 + z ** 2) < 1)]
z_1 = z[(np.sqrt(x ** 2 + y ** 2 + z ** 2) < 1)]

ax.scatter3D(x_1, y_1, z_1, color = 'purple', marker = '.', s = 10)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
import numpy as np
import matplotlib.pyplot as plt


ax = plt.axes(projection='3d')


r= np.random.uniform(0, 1, 5000)
phi = np.random.uniform(0, 2 * np.pi, 5000)


x = r * np.cos(phi)
y = r * np.sin(phi)
z = np.sqrt(1 - x ** 2 - y ** 2)
ax.scatter3D(x, y, z, color = 'purple', marker = '.', s = 2)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

### <strong>Distribución uniforme en un plano</strong>

De la misma forma que con el cubo, se generan puntos aleatorios con distribución uniforme y se grafican:

```python
x = np.random.uniform(-1, 1, 1000)
y = np.random.uniform(-1, 1, 1000)

plt.scatter(x, y, color = 'yellow', marker = '.')
plt.xlabel('x')
plt.ylabel('y')
```
