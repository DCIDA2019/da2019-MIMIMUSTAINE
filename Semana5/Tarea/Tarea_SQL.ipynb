# <strong>SQL</strong>

<strong>SQL</strong> es un lenguaje de programación dedicado al análisis de datos. Una de sus principals características es la facilidad con la que pueden filtrarse o seleccionarse datos por medio de etiquetas generalmente dadas en una columna.

También es posible utilizarlo para producir o modificar tablas de datos.

### <strong>Using SQL for Queries</strong>

#### <strong>LIMIT</strong>

Con esta etiqueta se especifica el número de datos que desea mostrarse, estos datos están sujetos a características compartidas.

```python
pdsql("SELECT * FROM Tmax_2018 LIMIT 4;") #Del archivo Tmax_2018 se muestran las primeras 4 filas.
```

#### <strong>WHERE</strong>

Con esta etiqueta es una característica compartida de los datos o es el conjunto de datos que cumplen con una condición, en este caso se cumple que la impresión de datos corresponde a aquellas ciudades que pertenecen al estado de Guanajuato y su Tmax > 25.0°C.

```python
#Mostrar primeros cinco elementos donde el estado sea Guanajuato.
pdsql("SELECT * FROM Tmax_2018 WHERE Edo = 'GTO' and Tmax > 25.0 LIMIT 5;")
```

#### <strong>SELECT</strong>

Con esta etiqueta se seleccionan las columnas que se desea imprimir.

```python
#Mostrar cinco primeros elementos de tres columnas: Lon, Lat, Tmax donde el estado sea Guanajuato.
pdsql("SELECT Lon, Lat, Tmax FROM Tmax_2018 WHERE Edo = 'GTO' and Tmax > 25.0 LIMIT 5;")
```

Note que los datos en los dos frames anteriores coinciden.

### <strong>Ordering</strong>

Para ordenar los datos de forma estructurada se hace uso de la etiqueta <strong>ORDER BY</strong>. Generalmente se ordena de forma ascendente, descendente.

```python
#Seleccione de Tmax_2018 donde el estado sea Guanajuato, ordenando ascendientemente Tmax, solo 5 elementos.
#Además Tmax tiene que ser mayor a 20°C.
etiqueta = """
SELECT * FROM Tmax_2018 
WHERE Edo = 'GTO' and Tmax > 25.0
ORDER BY Tmax ASC
LIMIT 5;
"""
pdsql(etiqueta)
```

### <strong>Aggregate queries</strong>

<strong>1. AS.</strong> Funciona para renombrar las columnas, es importante añadir comillas en nombres en forma de cadenas.

<strong>2. GROUP BY.</strong> Funciona para agrupar los datos, en este caso se agrupan de acuerdo a la columna Edo, donde aquellos que comparten una etiqueta, e.g., 'GTO' se promedian y se muestra sólo en una ocasión el valor de la temperatura.

<strong>3. HAVING.</strong> Una condición o característica más impuesta, tiene una función similar a <strong>WHERE</strong>.

```python
#Cambie el nombre de las columnas con el comando AS, desde el archivo Tmax_2018 donde Tmax in [10.0,20.0].
#Agrupe los datos por estado.
etiqueta = """
SELECT Lon AS Longitud, Lat AS Latitud, Edo AS Estado, Tmax AS 'Temperatura (°C)' FROM Tmax_2018
WHERE Tmax >10.0 and Tmax < 20.0
GROUP BY Edo
--HAVING Edo = 'GTO'
ORDER BY Tmax DESC
LIMIT 10;
"""
pdsql(etiqueta)
```

Es posible hacer uso de cualquier operador lógico, para comparar expresiones y realizar operaciones deseadas, algunos usos comunes es <strong>AND, OR, NOT</strong>.

## <strong>Joins</strong>

Se comparte información de dos o más archivos al relacionar el nombre o contenido de las columnas.

<strong>1. INNER JOINS.</strong> Se imprimen las intersecciones donde la condición <strong>ON</strong> se cumple. 

```python
etiqueta = """
SELECT a.Lat AS Latitud, a.Lon AS Longitud, a.Edo AS Estado, a.Tmax AS 'Temperatura máxima 2018 (°C)', 
b.Tmax AS 'Temperatura máxima 2019 (°C)'
FROM Tmax_2018 a
INNER JOIN Tmax_2019 b
ON a.Edo = b.Edo
GROUP BY a.Edo
LIMIT 10;
"""
pdsql(etiqueta)
```

### <strong>Outer join</strong>

<strong>2. LEFT OUTER JOIN.</strong> Se imprimen todos los elementos de comparación, en este caso la columna Est en a, b aunque en Temperatura 2018, Temperatura 2019 no exista información registrada.

```python
etiqueta = """
SELECT a.Lat AS Latitud, a.Lon AS Longitud, a.Est AS Ciudad, a.Tmax AS 'Temperatura máxima 2018 (°C)', 
b.Tmax AS 'Temperatura máxima 2019 (°C)'
FROM Tmax_2018 a
LEFT OUTER JOIN Tmax_2019 b
ON a.Est = b.Est
LIMIT 10;
"""
pdsql(etiqueta)
```

Con la figura anterior es posible notar que no todas las ciudades están dentro del contenido de Tmax_2018, Tmax_2019. Así, solo se muestran los datos en los que el nombre de la ciudad coincide.

<strong>3. UNION ALL.</strong> Se imprimen todos los datos en los que por lo menos en una columna exista información.

```python
#UNION ALL.
etiqueta = """
SELECT a.Lat AS Latitud, a.Lon AS Longitud, a.Est AS Ciudad, a.Tmax AS 'Temperatura máxima 2018 (°C)', 
b.Tmax AS 'Temperatura máxima 2019 (°C)'
FROM   Tmax_2018 a
       LEFT JOIN Tmax_2019 b
          ON a.Est = b.Est
UNION All
SELECT a.Lat AS Latitud, a.Lon AS Longitud, a.Est AS 'Ciudad 2018', a.Tmax AS 'Temperatura máxima 2018 (°C)', b.Tmax AS 'Temperatura máxima 2019 (°C)'
FROM   Tmax_2019 b
       LEFT JOIN Tmax_2018 a
          ON a.Est = b.Est
WHERE  a.Est IS NULL;
"""
pdsql(etiqueta)
```

En la tabla anterior se muestran todos los espacios en donde existen datos pero no se tiene una etiqueta común, en este caso la etiqueta que se busca común es el nombre de la ciudad.

<strong>4. Unión uno a uno</strong> Comparación para más de dos frames, se agrega cada elemento al cumplirse con condiciones impuestas. No es necesario de que todos los frames tengan condiciones entre todos.

Para el año 2019 se realiza básicamente el mismo método pero agregando información de más de dos Frames, se imponen condiciones comparando el primer archivo con los demás, se muestra la información donde ocurre intersección en la columna 'Est' en Tmax_2019.

```python
etiqueta = """
SELECT a.Lat, a.Lon, a.Est, a.Tmax, b.Tmed, c.Tmin
FROM Tmax_2019 a
INNER JOIN Tmed_2019 b ON a.Est = b.Est
INNER JOIN Tmin_2019 c ON c.Est = a.Est
LIMIT 10;
"""
pdsql(etiqueta)
```

<strong>Si requiere observar los datos en los cuales no hay intersección,</strong> de igual forma se puede hacer haciendo uso de <strong>LEFT OUTER JOIN</strong>. A continuación se muestra el mismo procedimiento:

```python
etiqueta = """
SELECT a.Lat, a.Lon, a.Est, a.Tmax, b.Tmed, c.Tmin
FROM Tmax_2019 a
LEFT OUTER JOIN Tmed_2019 b ON a.Est = b.Est
LEFT OUTER JOIN Tmin_2019 c ON c.Est = a.Est
LIMIT 10;
"""
pdsql(etiqueta)
```
