```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
from random import random
import scipy.stats
#import corner
from mpl_toolkits import mplot3d

plt.style.use('seaborn-whitegrid')

np.random.seed(2) #Número aleatorio para que los procesos se asimilen.
```

```python
path = '/home/mimimustaine/Downloads/cms-jupyter-materials-english-1.0/Data/'
v = pd.read_csv(path + 'Zmumu_Run2011A.csv')
v.head()
```

Para determinar la masa invariante se requiere de la siguiente expresión:

\begin{equation}
    M = \sqrt{2p_{T_{1}}p_{T_{2}}(cosh(\eta_{1} - \eta_{2}) - cos(\phi_{1} - \phi_{2}))},
\end{equation}

Se grafican histogramas de cada uno de los componentes para observar una relación entre \textbf{M} y los parámetros que se observan en la expresión anterior. 

El propósito es modelar <strong>M</strong> en términos de uno o alguno de estos parámetros.

```python
para = ['pt1', 'pt2', 'eta1', 'eta2', 'phi1', 'phi2']
color = ['red', 'purple', 'blue', 'pink', 'yellow', 'orange']

for i in range(len(para)):
    plt.hist(v[para[i]], 10, alpha = 0.6, color = color[i], label = 'Parámetro: {}'.format(para[i]))
    plt.legend(frameon = True, loc = 'best')
    plt.xlabel('{}'.format(para[i]))
    plt.ylabel('Frecuencia')
    plt.show()
```

Ahora se determinará <strong>M,</strong> después se producirá una función respecto a los parámetros y se observará dependencia para un modelado.

```python
M = np.sqrt(2 * v.pt1 * v.pt2 * (np.cosh(v.eta1 - v.eta2) - np.cos(v.phi1 - v.phi2)))
```

```python
for i in range(len(para)):
    #plt.plot(v[para[i]], M, color = color[i], label = '{}'.format(para[i]))
    plt.scatter(v[para[i]], M, color = color[i], alpha = 0.6, label = '{}'.format(para[i]))
    plt.legend(frameon = True, loc = 'best')
    plt.title('M vs {}'.format(para[i]))
    plt.xlabel('{}'.format(para[i]))
    plt.ylabel('M')
    plt.show()
```
