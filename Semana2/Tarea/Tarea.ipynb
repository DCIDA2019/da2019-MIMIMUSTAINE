### <strong>GroupBy: Split, Apply, Combine</strong>

Este conjunto de operaciones corresponde a las siguientes acciones:
<u1>
    <li><strong>Split.</strong> Divide al objeto y posteriormente lo agrupa en conjuntos con características compartidas.</li>
    <li><strong>Apply.</strong> Implica realizar una función sobre los conjuntos que pertenecen al objeto.</li>
    <li><strong>Combine.</strong> Combina el resultado de las operaciones anteriores en una matriz de salida.</li>
</u1>

```python
estados4
```

```python
#Existe una función predefinida en Python que permite obtener información general acerca del objeto.
estados4.groupby('Longitud').describe() #'describe()' proporciona información como el número de veces que se
                                        #repite un elemento, la media, desviación estándar, etc.
```

<strong>Aggregate, filter, transform, apply</strong>

```python
#----------aggregate()----------#
#Con esta función agregamos más información acerca del objeto, que no toda está contenida en la función
#describe().
estados4.groupby('Longitud').aggregate(['min', np.median, max])
```

```python
#----------filtering----------#
#Función que se ocupa de discriminar elementos de un objeto. En este caso, aquellos elementos en 'mLatitud'
#cuya media sea mayor a 22.0, puede comprobarse el resultado observando la tabla inicial.
def f(x):
    return x['mLatitud'].mean() > 22.0
estados4.groupby('mLatitud').filter(f)
```

```python
#----------transformation----------#
#La transformación, como su nombre lo especifica, tranforma cada una de las componentes.
estados4.Longitud
```

```python
#Note que la operación se realiza sobre las columnas diferentes de 'mLatitud'.
estados4.groupby('mLatitud').transform(lambda x : x ** 2)
```

```python
#----------apply() method----------#
#Se realiza una operación sobre el conjunto de datos seleccionados.
def norma(x): #Que el argumento sea un objeto tipo Series, DataFrame de Pandas.
    x['Longitud'] = x['Longitud'].sum()
    return x
#La función "normaliza el conjunto de datos seleccionados, en este caso la columna 'Longitud'".
estados4
```

```python
estados4.groupby('Longitud').apply(norma)
```

Anteriormente se realizaron operaciones sobre un conjunto de datos que compartían una posición, en los casos anteriores se trató de una columna común, ¿qué pasa cuando requiero que una función se ejecute en más de una columna (en este caso)?

### <strong>Pivot tables</strong>

Con funciones parecidas a las de <strong>Groupby, Pivot tables</strong> trabaja con una cadena de tal forma que la agrupa en un arreglo de dos o más dimensiones, el propósito es organizar la información.

```python
mergepivot1 = {1 : v1_1[0][0],
          2 : v1_1[1][0],
          3 : v1_1[2][0],
          4 : v1_1[3][0],
          5 : v1_1[4][0]}
longitudpivot = pd.Series(mergepivot1)
mergepivot2 = {1: v1_1[0][1],
          2 : v1_1[1][1],
          3 : v1_1[2][1],
          4 : v1_1[3][1],
          5 : v1_1[4][1]}
latitudpivot = pd.Series(mergepivot2)
mergepivot3 = {1 : v1_1[0][2],
          2 : v1_1[1][2],
          3 : v1_1[2][2],
          4 : v1_1[3][2],
          5 : v1_1[4][2]}
estadopivot = pd.Series(mergepivot3)
mergepivot4 = {1 : v1_1[0][3],
          2 : v1_1[1][3],
          3 : v1_1[2][3],
          4 : v1_1[3][3],
          5 : v1_1[4][3]}
clavepivot = pd.Series(mergepivot4)
mergepivot5 = {1 : v1_1[0][4],
          2 : v1_1[1][4],
          3 : v1_1[2][4],
          4 : v1_1[3][4],
          5 : v1_1[4][4]}
nombrepivot = pd.Series(mergepivot5)
mergepivot6 = {1 : v1_1[0][5],
          2 : v1_1[1][5],
          3 : v1_1[2][5],
          4 : v1_1[3][5],
          5 : v1_1[4][5]}
enepivot = pd.Series(mergepivot6)
estadospivot = pd.DataFrame({'Longitud': longitudpivot, 'Latitud': latitudpivot, 'Estado': estadopivot, 'Clave': clavepivot, 'Nombre': nombrepivot, 'ENE': enepivot})
estadospivot
```

```python
#----------Ejemplo de uso objetos 'pivot'----------
#Se introducirá en la tabla el valor de los elementos que existen en la columna 'Latitud', pero interesa
#conocer su valor en la intersección de 'Nombre' y 'Latitud'.
estadospivot.pivot_table('Longitud', index = 'Nombre', columns = 'Latitud')
```

```python
#----------Clases multivariadas----------
#Para especificar con mayor precisión la información de la que se trata en el contenido del cuadro original,
#se acotarán intervalos en el mismo.
rango = pd.cut(estadospivot['ENE'], [0, 50, 100]) #Intervalos de (0,50], (50, 100].
estadospivot.pivot_table('Longitud', ['Nombre', rango], 'Latitud')
```

```python
#PRODUCCIÓN DE TABLAS MULTIVARIADAS (más especificaciones), se genera otro intervalo en otra columna que
#contiene el objeto.
rango1 = pd.qcut(estadospivot['Latitud'], 2)
estadospivot.pivot_table('Longitud', ['Nombre', rango], ['Clave', rango1])
```

```python
#Es posible operar sobre los elementos que contiene el objeto nuevo.
estadospivot.pivot_table(index = 'Nombre', columns = 'Latitud', aggfunc = {'Longitud' : np.abs})
#Se hace uso de la función de valor absoluto en los elementos de 'Longitud'.
```

#### <strong>Resampling, Shifting, and Windowing</strong>

```python
#Ejercicio de producir un array con la fecha usando datos del archivo descargado en el inicio.
#Aquí se produce un algoritmo para asignar un día a cada uno de los elementos del arreglo.
fecha = []
aux = 20180101
i = 0
while i < 1884:
    for j in range(12):
        for k in range(28):
            if i == 1884:
                break
            fecha.append(aux + i)
            i += 1  
        aux += 72
    aux += 8801

vfecha = np.array(fecha)
```

```python
#Una vez producido el arreglo se convirtió a fecha y se produjo con la información un DataFrame.
nfecha = pd.to_datetime(fecha, format='%Y%m%d', errors='ignore')
datafecha = pd.DataFrame({'Fecha': nfecha, 'ENE': v1_1[:,5]})
datafecha
```

```python
#Se graficó la lluvia (se desconocen las unidades) por ciudad, se tienen 1883 ciudades etiquetadas con su
#número correspondiente.
c = datafecha['ENE']
c.plot(alpha=0.8, color = 'k', style=':')
#c.plot(style=':')
#c.plot(style='--')
plt.legend(['Lluvias por ciudad'],loc='upper right')
```

```python
#Desplazamiento en la misma figura.
fig, ax = plt.subplots(2, sharey=True)# apply a frequency to the data

c.plot(ax=ax[0], color = 'dimgray')
c.shift(900).plot(ax=ax[1], color = 'dimgray')

ax[0].legend(['Lluvias por estado'], loc=1)
ax[0].get_xticklabels()[4].set(weight='heavy', color='red')
ax[0].axvline(200, alpha=0.3, color='red')

ax[1].legend(['Desplazamiento en 900 iteraciones'], loc=1)
ax[1].get_xticklabels()[4].set(weight='heavy', color='red')
ax[1].axvline(1075, alpha=0.3, color='red')
```

### <strong>High-Performance Pandas: eval() and query()</strong>

En la función <strong>eval()</strong> en <i>Pandas</i> se hace uso de expresiones como cadenas para eficientar operaciones usando <i>DataFrames</i>. La función aplicada a alguna operación, una vez que se compruebe que es posible usarla es aproximadamente <i>50%</i> más rápida que una la ejecución directa de la operación.

```python
#----------Operaciones aritméticas----------#
pd.eval('estados1 + estados2')
```

```python
#----------Operaciones de comparación----------#
pd.eval('estados1 < estados2')
```

```python
#Dentro de los componentes de un objeto DataFrame:
estados1.eval('Diferencia = Longitud - Latitud', inplace = True)
estados1.head()
```

En la función <strong>DataFrame.query()</strong> también se trabaja con cadenas, tiene la misma función que .eval().
